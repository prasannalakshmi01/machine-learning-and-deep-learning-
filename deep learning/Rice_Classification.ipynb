### Importing required python libraries

```python
import os
import pathlib
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Get all the image files from a folder 

```python
paths=pathlib.Path(r"C:\Users\harsh\Downloads\Rice_Image_Dataset")
lst=list(paths.glob("*/*.jpg"))
lst
```

#### Converting all the files into paths using string function

```python
paths = []
for x in lst:
    paths.append(str(x))
```

```python
# Number of file paths
len(paths)
```

```python
# to get all the labels for the paths
labels=[]
for i in paths:
    folder_path=os.path.dirname("{}".format(i))
    labels.append(os.path.basename(folder_path))
```

```python
labels
```

```python
# Number of labels
len(labels)
```

```python
# Mapping categorical values to numerical values explicitly
label = {'Arborio' : 1, 'Basmati' : 2, 'Ipsala' : 3, 'Jasmine' : 4, 'Karacadag' : 0}
labels_=[]
for i in labels:
    i=label[i]
    labels_.append(i)
```

```python
labels_
```

```python
import numpy as np
pd.DataFrame(np.array(labels_)).value_counts()
```

```python
glob.glob(r"C:\Users\harsh\Downloads\Rice_Image_Dataset\{}/*.*".format("Ipsala"))[0]
```

#### Visualization of images of different rice categories

```python
import cv2
types = ["Arborio","Basmati","Ipsala","Jasmine","Karacadag"]
for i in types:
    plt.figure(figsize=(2,2))
    img=cv2.imread(glob.glob(r"C:\Users\harsh\Downloads\Rice_Image_Dataset\{}/*.*".format(i))[3])
    img=cv2.resize(img,(80,80))
    plt.ylabel(i)
    plt.imshow(img)
```

```python
# resizing all the images to a standard size of 80*80
data = []
for i in paths:
    img = cv2.imread(i)
    img = cv2.resize(img,(80,80))
    data.append(img)
```

```python
data
```

```python
len(data)
```

### Splitting of entire data into train and test 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(np.array(data),np.array(labels_),stratify=np.array(labels_),test_size=0.10,random_state=100)
```

```python
# shape of train and test data
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

##### importing libraries required for model training

```python
import keras
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Dense,Flatten
from keras.utils import np_utils
```

```python
pd.DataFrame(y_train).value_counts()
```

```python
# converting y_train data using one hot encoding
y_train_cat=keras.utils.to_categorical(y_train,dtype=int)
```

```python
y_train_cat
```

```python
# converting y_test data using one hot encoding
y_test_cat=keras.utils.to_categorical(y_test,dtype=int)
```

```python
y_train_cat.shape
```

### Generalized model Architecture

```python
model=Sequential()
model.add(Conv2D(32,(3,3),padding="same",activation="relu",input_shape=(80,80,3)))
model.add(MaxPooling2D(pool_size=(1,1)))

model.add(Conv2D(32,(3,3),padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=(1,1)))

model.add(Conv2D(64,(3,3),padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3),padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(128,(3,3),padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))


model.add(Conv2D(128,(3,3),padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(256,(3,3),padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))



model.add(Flatten())
model.add(Dense(128,activation="relu",kernel_initializer="he_uniform"))
model.add(Dense(5,activation="softmax"))
```

```python
# model compilation
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=['accuracy'])
```

```python
# model training
model.fit(x_train,y_train_cat,batch_size=100,epochs=2,validation_split=0.15)
```

```python
# model prediction
y_test_pred = model.predict(x_test)
y_test_pred
```

```python
# to get the label of test data from the probabilities
y_pred_classes=[np.argmax(i) for i in y_test_pred] 
```

```python
pd.DataFrame(y_pred_classes).value_counts()
```

```python
y_test
```

```python
# model evaluation
model.evaluate(x_test,y_test_cat)
```

```python
from keras.applications.vgg16 import VGG16
```

```python
main_=VGG16(include_top=False, weights='imagenet',input_shape=((80,80,3)))
```

```python
main_.summary()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten
```

```python
# Data is large and dissimilar
model1=Sequential()
model1.add(main_)
model1.add(Flatten())
model1.add(Dense(80,activation="relu"))
model1.add(Dense(5,activation="softmax"))
```

```python
model1.compile(optimizer="adam", loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model1.fit(x_train,y_train_cat,batch_size=100,epochs=2,validation_split=0.15)
```
